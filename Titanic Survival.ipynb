### Encoding categorical features

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False,handle_unknown="ignore")
df_encoded = pd.DataFrame(encoder.fit_transform(df[["Sex","Embarked"]]),
                          columns=encoder.get_feature_names_out(["Sex","Embarked"]))
df.drop(columns=["Sex","Embarked"],inplace=True)
df = pd.concat([df,df_encoded],axis=1)
df.head()
```

```python
df.info()
```

```python
df=df.drop("Name",axis=1)
```

```python
df = df.drop("Ticket",axis=1)
```

```python
df.shape
```

## Model training and prediction

```python
X = df.drop("Survived",axis=1)
y = df["Survived"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
```

```python
clf.fit(X_train,y_train);
```

```python
y_preds = clf.predict(X_test)
```

```python
clf.score(X_test,y_test) # achieving 100% accuracy
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_preds))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
```

```python
preds = model.predict(X_test)
```

```python
model.score(X_test,y_test) #achieving 100% accuracy
```

```python
print(classification_report(y_test,preds))
```
